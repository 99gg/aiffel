```python
# 도미 bream = 1, 방어 smelt = 0
fish_target = [1] * 35 + [0] * 14
print(fish_target)


from sklearn.neighbors import KNeighborsClassifier


kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
score = kn.score(fish_data, fish_target)
print(f"score: {score}")
pred = kn.predict([[30, 600], [30, 50], [1, 50]])
print(f"pred: {pred}")


print(kn._fit_X)
print(kn._y)
```

```python
scores = []
for n in range(5, 50):
    kn.n_neighbors = n
    score = kn.score(fish_data, fish_target)
    # print(score)
    if score < 1:
        print(n, score)
    scores.append([n, score])

plt.scatter([n for n, s in scores], [s for n, s in scores], marker="x")
plt.show()
```
